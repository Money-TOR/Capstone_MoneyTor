### Cleaning Data

```python
# =========================================================== #
# 1. Mengubah tipe data tanggal dari object menjadi datetime! #
# =========================================================== #

# Kolom yang akan dikonversi antara lain
#     transactions_df['tanggal']
#     sellers_df['tanggal_bergabung']
#     products_df['tanggal_input_produk']
#     logs_df['tanggal_log']

transactions_df['tanggal'] = pd.to_datetime(transactions_df['tanggal'])
sellers_df['tanggal_bergabung'] = pd.to_datetime(sellers_df['tanggal_bergabung'])
products_df['tanggal_input_produk'] = pd.to_datetime(products_df['tanggal_input_produk'])
logs_df['tanggal_log'] = pd.to_datetime(logs_df['tanggal_log'])

# Mengecek salah satu dataset yang sudah diubah kolom objectnya
transactions_df.info()
```

```python
# ========================== #
# 2. Menangani Missing Value #
# ========================== #

# Missing Value yang ingin ditangani berasal dari dataset Transactions
#     Kolom Event  : diisii dengan tidak ada event
#     Kolom Diskon : diisi dengan 0

transactions_df['event'] = transactions_df['event'].fillna('Tidak Ada Event')
transactions_df['diskon'] = transactions_df['diskon'].fillna(0)

# Mengecek status kolom pada dataset transactions
transactions_df.info()
```

```python
# ======================= #
# 3. Validasi Tidak Wajar #
# ======================= #

# Pada tahap assessing sebelumnya ditemukan: total_harga minimum = 0
# perlu memastikan apakah data tersebut: valid atau merupakan anomali/error.

# 3a. Mengecek transaksi dengan total harga = 0
transactions_df[transactions_df['total_harga'] == 0]
```

```python
# ======================= #
# 3. Validasi Tidak Wajar #
# ======================= #

# Pada tahap 3a sebelumnya terlihat bahwa:
# total_harga = 0 bukan error acak, karena seluruh transaksi tersebut memiliki: status_transaksi = batal

# 3b. Mengecek apakah semua total_harga == 0 memang memiliki status_transaksi == 'batal'
transactions_df[transactions_df['total_harga'] == 0]['status_transaksi'].value_counts()
```

```python
# ========================================= #
# 4. Mengecekk konsistensi Data Kategorikal #
# ========================================= #

# memastikan tidak ada penulisan kategori yang berbeda seperti:
#     Cash
#     cash
#     CASH

print("Metode Pembayaran:")
print(transactions_df['metode_pembayaran'].unique())

print("\nStatus Transaksi:")
print(transactions_df['status_transaksi'].unique())

print("\nJenis Transaksi:")
print(transactions_df['jenis_transaksi'].unique())

print("\nKategori Transaksi:")
print(transactions_df['kategori'].unique())

print("\nStatus Produk:")
print(products_df['status_produk'].unique())
```

```python
# ================================== #
# 5a.Validasi Perhitungan Total Harga #
# ================================== #

# memastikan rumus transaksi benar.
#    total_harga = qty × harga_satuan

invalid_total = transactions_df[
    (transactions_df['qty'] * transactions_df['harga_satuan'] != transactions_df['total_harga']) &
    (transactions_df['status_transaksi'] != 'batal')
]

invalid_total
```

```python
# ================================================= #
# 5b.Validasi Perhitungan Total Harga dengan diskon #
# ================================================= #

# Pada tahap 5a sebelumnya terlihat bahwa:
# qty x harga_satuan != total_harga yang terjadi pada 68 transaksi sukses.
# Artinya transaksi sudah memperhitungkan diskon.

# Validasi total_harga dengan memperhitungkan diskon
validasi_diskon = transactions_df[
    (
        (
            transactions_df['qty'] *
            transactions_df['harga_satuan'] *
            (1 - transactions_df['diskon'] / 100)
        ).round(0) != transactions_df['total_harga']
    )
    &
    (transactions_df['status_transaksi'] != 'batal')
]

validasi_diskon
```

## Exploratory Data Analysis (EDA)

### Hal yang diexplore adalah:
Pada tahap analisis ini akan berpusat pada hubungan timbal-balik antara Pemasaran dan Inventori:

Dari pemasaran (Pelanggan & Penjualan): Mencari tahu produk apa yang paling laku, kapan waktu ramai pembeli, dan taktik promosi/pembayaran apa yang paling disukai pelanggan untuk mendongkrak pendapatan.

Dari inventori (Stok & Operasional): Memastikan kelancaran pasokan agar tidak kehilangan potensi penjualan akibat kehabisan stok serta memantau kesehatan pengeluaran bisnis melalui efisiensi penyuplai.

```python
# Memastikan hanya menganalisis transaksi sukses
sukses_transaksi_df = transactions_df[transactions_df['status_transaksi'] == 'sukses'].copy()

print(f"Total data transaksi berhasil: {len(sukses_transaksi_df)} baris")
```

```python
# Total Pemasukan & Pengeluaran
revenue_vs_expense = sukses_transaksi_df.groupby('jenis_transaksi')['total_harga'].sum().reset_index()
print("Total Pemasukan & Pengeluaran (Sukses)")
print(revenue_vs_expense)
print("\n")

# Rasio Transaksi Sukses & Batal
status_ratio = transactions_df['status_transaksi'].value_counts(normalize=True) * 100
print("Rasio Status Transaksi (%)")
print(status_ratio.round(2))
print("\n")

# Transaksi Berdasarkan Kategori
category_analysis = sukses_transaksi_df.groupby('kategori').agg(
    jumlah_transaksi=('id_transaksi', 'count'),
    total_nilai=('total_harga', 'sum')
).sort_values(by='total_nilai', ascending=False)
print("Analisis Berdasarkan Kategori Transaksi")
print(category_analysis)
```

```python
#  Menampilkan Top 5 Produk Terlaris berdasarkan Kuantitas (Qty)
top_products_qty = sukses_transaksi_df.groupby('id_produk')['qty'].sum().reset_index()
top_products_qty = top_products_qty.merge(products_df[['id_produk', 'nama_produk', 'kategori_produk']], on='id_produk')
top_products_qty = top_products_qty.sort_values(by='qty', ascending=False).head(5)

print("Top 5 Produk Terlaris (Kuantitas)")
print(top_products_qty[['nama_produk', 'kategori_produk', 'qty']].to_string(index=False))
print("\n")

# Metode Pembayaran Terpopuler
payment_trends = sukses_transaksi_df.groupby('metode_pembayaran').agg(
    jumlah_pengguna=('id_transaksi', 'count'),
    total_nominal=('total_harga', 'sum')
).sort_values(by='jumlah_pengguna', ascending=False)

print("Preferensi Metode Pembayaran")
print(payment_trends)
print("\n")

# Pengaruh Diskon terhadap Pembelian Kuantitas (Qty)
promo_impact = sukses_transaksi_df.groupby('event').agg(
    rata_rata_qty=('qty', 'mean'),
    rata_rata_diskon=('diskon', 'mean'),
    total_omset=('total_harga', 'sum')
).sort_values(by='total_omset', ascending=False)

print("Dampak Event Promosi & Diskon")
print(promo_impact)
```

```python
# Memastikan kolom tanggal bertipe datetime
sukses_transaksi_df['tanggal'] = pd.to_datetime(sukses_transaksi_df['tanggal'])

# Analisis Hari Ter ramai dalam Seminggu
sukses_transaksi_df['hari_pembelian'] = sukses_transaksi_df['tanggal'].dt.day_name()
order_hari = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

traffic_hari = sukses_transaksi_df.groupby('hari_pembelian').agg(
    jumlah_transaksi=('id_transaksi', 'count'),
    total_omset=('total_harga', 'sum')
).reindex(order_hari)

print("Analisis Trafik & Penjualan Berdasarkan Hari")
print(traffic_hari)
```

```python
#  Hitung total produk terjual dari transaksi sukses
total_terjual = sukses_transaksi_df.groupby('id_produk')['qty'].sum().reset_index()
total_terjual.columns = ['id_produk', 'total_terjual']

# Hitung Sisa Stok Aktual (Stok Awal - Total Terjual)
stok_aktual_df = products_df.merge(total_terjual, on='id_produk', how='left')
stok_aktual_df['total_terjual'] = stok_aktual_df['total_terjual'].fillna(0)
stok_aktual_df['stok_sekarang'] = stok_aktual_df['stok_awal'] - stok_aktual_df['total_terjual']

# Deteksi Produk yang Mendekati/Di Bawah Batas Minimum Stok
urgent_restock_aktual = stok_aktual_df[stok_aktual_df['stok_sekarang'] <= stok_aktual_df['minimum_stok']]

print(f"Deteksi Kritis: Ada {len(urgent_restock_aktual)} Produk Butuh Restock Segera")
if len(urgent_restock_aktual) > 0:
    print(urgent_restock_aktual[['nama_produk', 'stok_awal', 'total_terjual', 'stok_sekarang', 'minimum_stok']].to_string(index=False))
else:
    print("Semua stok produk aman di atas batas minimum.")
```

```python
# Menghubungkan top 5 produk terlaris dengan sisa stok real-time saat ini
cross_analysis = top_products_qty[['id_produk', 'nama_produk', 'qty']].merge(
    stok_aktual_df[['id_produk', 'stok_sekarang', 'minimum_stok']],
    on='id_produk'
)

print("Analisis Lintas Divisi: Kesehatan Stok untuk Produk Terlaris")
print(cross_analysis.to_string(index=False))
print("\n")

# Performa Penjualan Berdasarkan Nama Usaha Seller
seller_performance = sukses_transaksi_df.merge(sellers_df, on='id_seller')
seller_summary = seller_performance.groupby('nama_usaha').agg(
    total_penjualan=('total_harga', 'sum'),
    jumlah_transaksi=('id_transaksi', 'count')
).sort_values(by='total_penjualan', ascending=False)

print("Performa Penjualan Berdasarkan Nama Usaha")
print(seller_summary)
```

Batas minimum stock harus dinaikkan 1.5x lipat khusus untuk produk Sembako dan Kopi sebulan sebelum event dimulai sebagai stok pengaman.

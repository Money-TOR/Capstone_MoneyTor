### Cleaning Data

```python
# =========================================================== #
# 1. Mengubah tipe data tanggal dari object menjadi datetime! #
# =========================================================== #

# Kolom yang akan dikonversi antara lain
#     transactions_df['tanggal']
#     sellers_df['tanggal_bergabung']
#     products_df['tanggal_input_produk']
#     logs_df['tanggal']

transactions_df['tanggal'] = pd.to_datetime(transactions_df['tanggal'])
sellers_df['tanggal_bergabung'] = pd.to_datetime(sellers_df['tanggal_bergabung'])
products_df['tanggal_input_produk'] = pd.to_datetime(products_df['tanggal_input_produk'])
logs_df['tanggal'] = pd.to_datetime(logs_df['tanggal'])

# Mengecek salah satu dataset yang sudah diubah kolom objectnya
transactions_df.info()
```

```python
# ================================ #
# 2. Menangani format kolom Diskon #
# ================================ #

# Diskon masih bertipe object dengan format campuran sebagai berikut:
#     "Harga Normal"
#     "Diskon 15%"
#     "Diskon 20%"

# Menghapus teks pada kolom diskon
transactions_df['diskon'] = (
    transactions_df['diskon']
    .str.replace('Diskon ', '', regex=False)
    .str.replace('%', '', regex=False)
    .replace('Harga Normal', '0')
)

# Mengubah tipe data menjadi integer yang sebelumnya object
transactions_df['diskon'] = (
    transactions_df['diskon']
    .astype(int)
)

# Mengecek status kolom diskon pada dataset transactions
transactions_df[['diskon']].info()

# Melihat nilai unik kolom diskon setelah cleaning
print(transactions_df['diskon'].unique())
```

```python
# =============================================== #
# 3. Validasi konsistensi perhitungan total_harga #
# =============================================== #

# Memastikan nilai total_harga pada dataset transaksi sudah sesuai,
# Mengecek apakah perhitungan transaksi benar setelah memperhitungkan diskon.

# Menghitung ulang total transaksi
cek_total = transactions_df[
    (
        (
            transactions_df['qty']
            * transactions_df['harga_satuan']
            * (1 - transactions_df['diskon'] / 100)
        ).round(0)
        != transactions_df['total_harga']
    )
]

# Menampilkan hasil
cek_total
```

```python
# ========================================= #
# 4. Mengecekk konsistensi Data Kategorikal #
# ========================================= #

# memastikan tidak ada penulisan kategori yang berbeda seperti:
#     Cash
#     cash
#     CASH

print("Metode Pembayaran:")
print(transactions_df['metode_pembayaran'].unique())

print("\nStatus Transaksi:")
print(transactions_df['status_transaksi'].unique())

print("\nJenis Transaksi:")
print(transactions_df['jenis_transaksi'].unique())

print("\nKategori:")
print(transactions_df['kategori'].unique())

print("\nEvent:")
print(transactions_df['event'].unique())

print("\nStatus Produk:")
print(products_df['status_produk'].unique())

print("\nJenis Perubahan:")
print(logs_df['jenis_perubahan'].unique())
```

```python
# ============================== #
# 5. Validasi log perubahan stok #
# ============================== #

# Memastikan perubahan stok masuk akal

cek_penjualan = logs_df[
    (logs_df['jenis_perubahan'] == 'Penjualan') &
    (
        logs_df['stok_sebelum']
        - logs_df['jumlah']
        != logs_df['stok_sesudah']
    )
]

cek_penjualan
```

### Insight Cleaning Data

- Pada tahap cleaning data dilakukan konversi tipe data pada seluruh atribut bertipe tanggal agar dapat digunakan untuk analisis berbasis waktu dan tren transaksi.

- Kolom diskon yang sebelumnya berbentuk teks berhasil ditransformasikan menjadi format numerik sehingga mempermudah proses perhitungan dan visualisasi.

- Dilakukan validasi terhadap perhitungan `total_harga` menggunakan kombinasi `qty`, `harga_satuan`, dan `diskon`, dan tidak ditemukan ketidaksesuaian pada seluruh data transaksi.

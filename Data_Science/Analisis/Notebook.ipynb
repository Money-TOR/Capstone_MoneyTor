### Cleaning Data

```python
# Mengubah tipe data tanggal dari object menjadi datetime! #


# Kolom yang akan dikonversi antara lain
transactions_df['tanggal'] = pd.to_datetime(transactions_df['tanggal'])
sellers_df['tanggal_bergabung'] = pd.to_datetime(sellers_df['tanggal_bergabung'])
products_df['tanggal_input_produk'] = pd.to_datetime(products_df['tanggal_input_produk'])
logs_df['tanggal'] = pd.to_datetime(logs_df['tanggal'])

# Mengecek salah satu dataset yang sudah diubah kolom objectnya
transactions_df.info()
```

```python
# Menangani format kolom Diskon

# Menghapus teks pada kolom diskon
transactions_df['diskon'] = (
    transactions_df['diskon']
    .str.replace('Diskon ', '', regex=False)
    .str.replace('%', '', regex=False)
    .replace('Harga Normal', '0')
)

# Mengubah tipe data menjadi integer yang sebelumnya object
transactions_df['diskon'] = (
    transactions_df['diskon']
    .astype(int)
)

# Mengecek status kolom diskon pada dataset transactions
transactions_df[['diskon']].info()

# Melihat nilai unik kolom diskon setelah cleaning
print(transactions_df['diskon'].unique())
```

```python
# Validasi konsistensi perhitungan total_harga

# Menghitung ulang total transaksi
cek_total = transactions_df[
    (
        (
            transactions_df['qty']
            * transactions_df['harga_satuan']
            * (1 - transactions_df['diskon'] / 100)
        ).round(0)
        != transactions_df['total_harga']
    )
]

# Menampilkan hasil
cek_total
```

```python
# Mengecekk konsistensi Data Kategorikal

# memastikan tidak ada penulisan kategori yang berbeda
print("Metode Pembayaran:")
print(transactions_df['metode_pembayaran'].unique())

print("\nStatus Transaksi:")
print(transactions_df['status_transaksi'].unique())

print("\nJenis Transaksi:")
print(transactions_df['jenis_transaksi'].unique())

print("\nKategori:")
print(transactions_df['kategori'].unique())

print("\nEvent:")
print(transactions_df['event'].unique())

print("\nStatus Produk:")
print(products_df['status_produk'].unique())

print("\nJenis Perubahan:")
print(logs_df['jenis_perubahan'].unique())
```

```python
# Validasi log perubahan stok

# Memastikan perubahan stok masuk akal

cek_penjualan = logs_df[
    (logs_df['jenis_perubahan'] == 'Penjualan') &
    (
        logs_df['stok_sebelum']
        - logs_df['jumlah']
        != logs_df['stok_sesudah']
    )
]

cek_penjualan
```

## Visualization & Explanatory Analysis

### Pertanyaan bisnis 1: Bagaimana tren total pendapatan transaksi UMKM dari tahun 2024 - 2026 dan kapan peningkatan transaksi tertinggi terjadi?

```python
# Filter hanya transaksi berhasil
revenue_df = transactions_df[
    transactions_df['status_transaksi'] == 'Berhasil'
].copy()

# Membuat kolom tahun
revenue_df['tahun'] = revenue_df['tanggal'].dt.year

# Menghitung total pendapatan per tahun
yearly_revenue = (
    revenue_df
    .groupby('tahun')['total_harga']
    .sum()
    .reset_index()
)

# Menghitung persentase pertumbuhan tahunan
yearly_revenue['growth_%'] = (
    yearly_revenue['total_harga']
    .pct_change() * 100
)

# Menampilkan data
display(yearly_revenue)

# Visualisasi tren pendapatan tahunan
plt.figure(figsize=(8,5))

plt.plot(
    yearly_revenue['tahun'],
    yearly_revenue['total_harga'],
    marker='o'
)

plt.title('Tren Total Pendapatan UMKM Tahun 2024–2026')
plt.xlabel('Tahun')
plt.ylabel('Total Pendapatan')
plt.xticks(yearly_revenue['tahun'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Insight pertanyaan bisnis 1
- Berdasarkan visualisasi, total pendapatan transaksi UMKM mengalami tren peningkatan dari tahun 2024 hingga 2026. Pendapatan meningkat cukup signifikan dari sekitar 274 juta pada tahun 2024 menjadi sekitar 892 juta pada tahun 2025, kemudian kembali meningkat menjadi lebih dari 1 miliar pada tahun 2026.

- Peningkatan pendapatan tertinggi terjadi pada periode 2024 ke 2025. Hal ini menunjukkan bahwa aktivitas transaksi dan penjualan UMKM berkembang secara positif selama periode pengamatan.

### Pertanyaan bisnis 2: Kategori produk atau layanan apa yang menghasilkan total pendapatan tertinggi selama periode 2025–2026 dan bagaimana perkembangan pendapatannya?

```python
# Filter transaksi berhasil
category_df = transactions_df[
    transactions_df['status_transaksi'] == 'Berhasil'
].copy()

# Membuat kolom tahun
category_df['tahun'] = category_df['tanggal'].dt.year

# Filter hanya tahun 2025–2026
category_df = category_df[
    category_df['tahun'].isin([2025, 2026])
]

# Total pendapatan per kategori
revenue_category = (
    category_df
    .groupby('kategori')['total_harga']
    .sum()
    .sort_values(ascending=False)
)

# Visualisasi kategori dengan pendapatan tertinggi
plt.figure(figsize=(8,5))
revenue_category.sort_values().plot(kind='barh')
plt.title('Total Pendapatan per Kategori (2025–2026)')
plt.xlabel('Total Pendapatan')
plt.ylabel('Kategori')
plt.grid(True)
plt.tight_layout()
plt.show()

print("\n")

# Tren pendapatan kategori per tahun
category_trend = (
    category_df
    .groupby(['tahun', 'kategori'])['total_harga']
    .sum()
    .unstack()
)

# Visualisasi tren kategori
category_trend.plot(figsize=(10,5), marker='o')
plt.title('Tren Pendapatan Kategori Tahun 2025–2026')
plt.xlabel('Tahun')
plt.ylabel('Total Pendapatan')
plt.xticks([2025, 2026])
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Pertanyaan bisnis 3: Bagaimana pengaruh event promosi terhadap nilai transaksi yang berhasil selama periode 2024 - 2026?

```python
# Filter transaksi berhasil
event_df = transactions_df[
    transactions_df['status_transaksi'] == 'Berhasil'
].copy()

# Total pendapatan berdasarkan event
event_revenue = (
    event_df
    .groupby('event')['total_harga']
    .sum()
    .sort_values(ascending=False)
)

# Visualisasi total pendapatan per event
plt.figure(figsize=(10,6))
event_revenue.sort_values().plot(kind='barh')
plt.title('Total Pendapatan Berdasarkan Event Promosi')
plt.xlabel('Total Pendapatan')
plt.ylabel('Event')
plt.grid(True)
plt.tight_layout()
plt.show()

print("\n")

# Rata-rata transaksi per event
avg_event = (
    event_df
    .groupby('event')['total_harga']
    .mean()
    .sort_values(ascending=False)
)

# Visualisasi rata-rata transaksi
plt.figure(figsize=(10,6))
avg_event.sort_values().plot(kind='barh')
plt.title('Rata-rata Nilai Transaksi per Event')
plt.xlabel('Rata-rata Total Transaksi')
plt.ylabel('Event')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Insight pertanyaan bisnis 3
- Berdasarkan visualisasi, sebagian besar pendapatan transaksi berasal dari kondisi Tidak Ada Event, yang menunjukkan bahwa aktivitas transaksi normal masih menjadi sumber utama pendapatan UMKM selama periode 2024 - 2026.

- Namun, beberapa event promosi seperti Ramadan dan Lebaran tetap memberikan kontribusi pendapatan yang cukup besar dibanding event lainnya. Selain itu, berdasarkan rata-rata nilai transaksi, event seperti 10.10 Sale, 8.8 Sale, dan 3.3 Sale menghasilkan rata-rata transaksi yang lebih tinggi dibanding event lainnya.

- Hal ini menunjukkan bahwa meskipun volume transaksi terbesar terjadi di luar periode promosi, beberapa event tertentu mampu mendorong pelanggan melakukan transaksi dengan nilai yang lebih besar.

### Pertanyaan bisnis 4: Metode pembayaran apa yang paling sering digunakan dan menghasilkan nilai transaksi terbesar pada tahun 2026?

```python
# Filter transaksi berhasil tahun 2026
payment_df = transactions_df[
    (transactions_df['status_transaksi'] == 'Berhasil') &
    (transactions_df['tanggal'].dt.year == 2026)
].copy()

# Frekuensi penggunaan metode pembayaran
payment_count = (
    payment_df['metode_pembayaran']
    .value_counts()
)

# Visualisasi frekuensi pembayaran
plt.figure(figsize=(7,5))
payment_count.sort_values().plot(kind='barh')
plt.title('Frekuensi Metode Pembayaran Tahun 2026')
plt.xlabel('Jumlah Transaksi')
plt.ylabel('Metode Pembayaran')
plt.grid(True)
plt.tight_layout()
plt.show()
print("\n")

# Total transaksi per metode pembayaran
payment_revenue = (
    payment_df
    .groupby('metode_pembayaran')['total_harga']
    .sum()
    .sort_values(ascending=False)
)

# Visualisasi total transaksi pembayaran
plt.figure(figsize=(7,5))
payment_revenue.sort_values().plot(kind='barh')
plt.title('Total Pendapatan Berdasarkan Metode Pembayaran Tahun 2026')
plt.xlabel('Total Pendapatan')
plt.ylabel('Metode Pembayaran')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Insight pertanyaan bisnis 4
- Berdasarkan visualisasi, metode pembayaran QRIS menjadi metode pembayaran yang paling sering digunakan selama tahun 2026 dengan jumlah transaksi tertinggi dibanding metode pembayaran lainnya. Selain itu, QRIS juga menghasilkan total pendapatan terbesar dibandingkan metode pembayaran Cash maupun Transfer.

- Sementara itu, metode pembayaran Cash berada pada posisi kedua baik dari jumlah transaksi maupun total pendapatan, sedangkan Transfer menjadi metode pembayaran dengan penggunaan dan kontribusi pendapatan paling rendah. Hal ini menunjukkan bahwa pelanggan cenderung lebih memilih metode pembayaran digital seperti QRIS dalam melakukan transaksi UMKM selama periode pengamatan.

## Konversi Tanggal dari Dataset

```python
# Konversi tanggal, tanggal input produk, dan tanggal bergabung
main_data['tanggal'] = pd.to_datetime(main_data['tanggal'])

main_data['tanggal_input_produk'] = pd.to_datetime(
    main_data['tanggal_input_produk']
)

main_data['tanggal_bergabung'] = pd.to_datetime(
    main_data['tanggal_bergabung']
)

# Cek Dataset setelah diubah
main_data.info()
```

## FEATURE ENGINEERING

Pada tahap feature engineering dilakukan pembuatan beberapa fitur baru untuk memperkaya informasi pada dataset transaksi UMKM. Fitur-fitur tambahan ini dibuat untuk membantu proses analisis, dashboard interaktif, serta mempermudah interpretasi pola transaksi, profitabilitas, perilaku pelanggan, dan kondisi inventori.


```python
# Feature Engineering 1
#    Pada tahap ini dibuat fitur profit untuk menghitung estimasi keuntungan dari
#    setiap transaksi berdasarkan selisih antara harga jual dan harga modal yang
#    dikalikan dengan jumlah produk terjual (qty). Fitur ini digunakan untuk
#    mengetahui keuntungan aktual dari aktivitas transaksi UMKM.

main_data['profit'] = (
    (main_data['harga_satuan'] - main_data['harga_modal'])
    * main_data['qty']
)
```

```python
# Feature Engineering 2
#    Fitur profit_margin_% dibuat untuk mengetahui persentase keuntungan dari
#    setiap produk terhadap harga jualnya. Feature ini membantu dalam menganalisis
#    efisiensi dan profitabilitas produk yang dijual.

main_data['profit_margin_%'] = (
    (
        main_data['harga_satuan']
        - main_data['harga_modal']
    )
    / main_data['harga_satuan']
) * 100
```

```python
# Feature Engineering 3
#    Fitur nama_hari dibuat dari kolom tanggal transaksi untuk mengetahui hari
#    terjadinya transaksi. Feature ini digunakan untuk menganalisis pola transaksi
#    harian selama periode pengamatan.

main_data['nama_hari'] = (
    main_data['tanggal']
    .dt.day_name()
)
```

```python
# Feature Engineering 4
#    Fitur jenis_hari dibuat untuk mengelompokkan transaksi ke dalam kategori
#    Weekend dan Weekday. Feature ini bertujuan untuk melihat perbedaan aktivitas
#    transaksi antara hari kerja dan akhir pekan.

main_data['jenis_hari'] = (
    main_data['nama_hari']
    .apply(
        lambda x:
        'Weekend'
        if x in ['Saturday', 'Sunday']
        else 'Weekday'
    )
)
```

```python
# Feature Engineering 5
#    Fitur jam dibuat dengan mengambil informasi jam dari kolom jam_transaksi.
#    Feature ini digunakan untuk membantu analisis waktu operasional dan pola
#    transaksi berdasarkan jam tertentu.

main_data['jam'] = pd.to_datetime(
    main_data['jam_transaksi'].astype(str),
    format='%H:%M'
).dt.hour
```

```python
# Feature Engineering 6
#    Fitur kategori_waktu dibuat untuk mengelompokkan transaksi berdasarkan
#    periode waktu seperti pagi, siang, sore, dan malam. Feature ini membantu
#    memahami waktu aktivitas transaksi pelanggan yang paling dominan.

def kategori_waktu(jam):

    if 5 <= jam < 12:
        return 'Pagi'

    elif 12 <= jam < 15:
        return 'Siang'

    elif 15 <= jam < 18:
        return 'Sore'

    else:
        return 'Malam'

main_data['kategori_waktu'] = (
    main_data['jam']
    .apply(kategori_waktu)
)
```

```python
# Feature Engineering 7
#    Fitur status_promo dibuat untuk membedakan transaksi yang menggunakan event
#    promosi dan transaksi normal tanpa promosi. Feature ini digunakan untuk
#    menganalisis pengaruh promosi terhadap aktivitas transaksi.

main_data['status_promo'] = (
    main_data['event']
    .apply(
        lambda x:
        'Promo'
        if x != 'Tidak Ada Event'
        else 'Tanpa Promo'
    )
)
```

```python
# Feature Engineering 8
#    Fitur rasio_stok dan kategori_stok dibuat untuk mengetahui kondisi persediaan
#    produk berdasarkan perbandingan antara stok awal dan minimum stok. Feature ini
#    membantu dalam memantau tingkat keamanan stok produk pada operasional UMKM.

main_data['rasio_stok'] = (
    main_data['stok_awal']
    / main_data['minimum_stok']
)

def kategori_stok(rasio):

    if rasio <= 1:
        return 'Kritis'

    elif rasio <= 2:
        return 'Rendah'

    elif rasio <= 4:
        return 'Aman'

    else:
        return 'Berlebih'

main_data['kategori_stok'] = (
    main_data['rasio_stok']
    .apply(kategori_stok)
)
```

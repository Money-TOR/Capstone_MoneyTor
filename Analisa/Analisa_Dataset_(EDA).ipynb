### Cleaning Data

```python
# =========================================================== #
# 1. Mengubah tipe data tanggal dari object menjadi datetime! #
# =========================================================== #

# Kolom yang akan dikonversi antara lain
#     transactions_df['tanggal']
#     sellers_df['tanggal_bergabung']
#     products_df['tanggal_input_produk']
#     logs_df['tanggal']

transactions_df['tanggal'] = pd.to_datetime(transactions_df['tanggal'])
sellers_df['tanggal_bergabung'] = pd.to_datetime(sellers_df['tanggal_bergabung'])
products_df['tanggal_input_produk'] = pd.to_datetime(products_df['tanggal_input_produk'])
logs_df['tanggal'] = pd.to_datetime(logs_df['tanggal'])

# Mengecek salah satu dataset yang sudah diubah kolom objectnya
transactions_df.info()
```

```python
# ================================ #
# 2. Menangani format kolom Diskon #
# ================================ #

# Diskon masih bertipe object dengan format campuran sebagai berikut:
#     "Harga Normal"
#     "Diskon 15%"
#     "Diskon 20%"

# Menghapus teks pada kolom diskon
transactions_df['diskon'] = (
    transactions_df['diskon']
    .str.replace('Diskon ', '', regex=False)
    .str.replace('%', '', regex=False)
    .replace('Harga Normal', '0')
)

# Mengubah tipe data menjadi integer yang sebelumnya object
transactions_df['diskon'] = (
    transactions_df['diskon']
    .astype(int)
)

# Mengecek status kolom diskon pada dataset transactions
transactions_df[['diskon']].info()

# Melihat nilai unik kolom diskon setelah cleaning
print(transactions_df['diskon'].unique())
```

```python
# =============================================== #
# 3. Validasi konsistensi perhitungan total_harga #
# =============================================== #

# Memastikan nilai total_harga pada dataset transaksi sudah sesuai,
# Mengecek apakah perhitungan transaksi benar setelah memperhitungkan diskon.

# Menghitung ulang total transaksi
cek_total = transactions_df[
    (
        (
            transactions_df['qty']
            * transactions_df['harga_satuan']
            * (1 - transactions_df['diskon'] / 100)
        ).round(0)
        != transactions_df['total_harga']
    )
]

# Menampilkan hasil
cek_total
```

```python
# ========================================= #
# 4. Mengecekk konsistensi Data Kategorikal #
# ========================================= #

# memastikan tidak ada penulisan kategori yang berbeda seperti:
#     Cash
#     cash
#     CASH

print("Metode Pembayaran:")
print(transactions_df['metode_pembayaran'].unique())

print("\nStatus Transaksi:")
print(transactions_df['status_transaksi'].unique())

print("\nJenis Transaksi:")
print(transactions_df['jenis_transaksi'].unique())

print("\nKategori:")
print(transactions_df['kategori'].unique())

print("\nEvent:")
print(transactions_df['event'].unique())

print("\nStatus Produk:")
print(products_df['status_produk'].unique())

print("\nJenis Perubahan:")
print(logs_df['jenis_perubahan'].unique())
```

```python
# ============================== #
# 5. Validasi log perubahan stok #
# ============================== #

# Memastikan perubahan stok masuk akal

cek_penjualan = logs_df[
    (logs_df['jenis_perubahan'] == 'Penjualan') &
    (
        logs_df['stok_sebelum']
        - logs_df['jumlah']
        != logs_df['stok_sesudah']
    )
]

cek_penjualan
```

## Exploratory Data Analysis (EDA)

### Hal yang diexplore adalah:
EDA Transaksi UMKM (2024-2026)
Mengeksplorasi data transaksi menganalisis tren pendapatan bulanan, mendeteksi kategori produk penyumbang pemasukan terbesar, mengukur efektivitas event promosi terhadap transaksi sukses, serta mengidentifikasi metode pembayaran yang paling populer dan menghasilkan nominal tertinggi.


```python
# Mengatur tema visualisasi agar rapi
sns.set_theme(style="whitegrid")

# Filter data hanya untuk transaksi pemasukan
pemasukan_df = transactions_df[transactions_df['jenis_transaksi'] == 'Pemasukan'].copy()

# Membuat kolom baru untuk Tahun-Bulan untuk melihat tren bulanan
pemasukan_df['tahun_bulan'] = pemasukan_df['tanggal'].dt.to_period('M')

# Agregasi total pendapatan per bulan
tren_pendapatan = pemasukan_df.groupby('tahun_bulan')['total_harga'].sum().reset_index()
tren_pendapatan['tahun_bulan'] = tren_pendapatan['tahun_bulan'].astype(str)

# Visualisasi Tren Pendapatan
plt.figure(figsize=(14, 6))
sns.lineplot(data=tren_pendapatan, x='tahun_bulan', y='total_harga', marker='o', color='teal', linewidth=2.5)
plt.title('Tren Total Pendapatan Bulanan UMKM (2024 - 2026)', fontsize=14, fontweight='bold')
plt.xlabel('Bulan-Tahun', fontsize=12)
plt.ylabel('Total Pendapatan (Rupiah)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Menampilkan dataframe hasil agregasi untuk insight teks
print("Tabel Tren Pendapatan 5 Bulan Teratas:")
print(tren_pendapatan.sort_values(by='total_harga', ascending=False).head())
```

```python
# Agregasi total pendapatan dan total kuantitas per kategori produk
kategori_revenue = pemasukan_df.groupby('kategori').agg(
    total_pendapatan=('total_harga', 'sum'),
    total_kuantitas=('qty', 'sum')
).reset_index().sort_values(by='total_pendapatan', ascending=False)

# Visualisasi Kategori Produk Tertinggi
plt.figure(figsize=(12, 6))
sns.barplot(data=kategori_revenue, x='total_pendapatan', y='kategori', palette='Blues_r')
plt.title('Total Pendapatan Berdasarkan Kategori Produk (2024 - 2026)', fontsize=14, fontweight='bold')
plt.xlabel('Total Pendapatan (Rupiah)', fontsize=12)
plt.ylabel('Kategori Produk', fontsize=12)
plt.tight_layout()
plt.show()

# Menampilkan dataframe hasil agregasi
print("Tabel Kontribusi Kategori Produk:")
print(kategori_revenue)
```

```python
# Filter data hanya untuk transaksi yang BERHASIL dan merupakan PEMASUKAN
transaksi_berhasil = pemasukan_df[pemasukan_df['status_transaksi'] == 'Berhasil'].copy()

# Membuat indikator apakah transaksi menggunakan promo atau tidak
transaksi_berhasil['pake_promo'] = transaksi_berhasil['event'].apply(lambda x: 'Tanpa Promo' if x == 'Tidak Ada Event' else 'Dengan Promo')

# 3a. Analisis Rata-rata Nilai Transaksi (Average Order Value)
aov_promo = transaksi_berhasil.groupby('pake_promo')['total_harga'].mean().reset_index()

# 3b. Analisis Total Nilai Transaksi per Event Promosi tertentu
event_revenue = transaksi_berhasil.groupby('event')['total_harga'].sum().reset_index().sort_values(by='total_harga', ascending=False)

# Visualisasi Distribusi dan Perbandingan AOV (Boxplot & Barplot)
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Barplot Rata-rata Nilai Transaksi
sns.barplot(data=aov_promo, x='pake_promo', y='total_harga', ax=ax[0], palette='Set2')
ax[0].set_title('Rata-rata Nilai per Transaksi (AOV)', fontsize=12, fontweight='bold')
ax[0].set_ylabel('Rata-rata Nominal (Rupiah)')
ax[0].set_xlabel('')

# Barplot Total Pendapatan per Event Promo (Top 5 Event Terbesar di luar 'Tidak Ada Event')
top_event = event_revenue[event_revenue['event'] != 'Tidak Ada Event'].head(5)
sns.barplot(data=top_event, x='total_harga', y='event', ax=ax[1], palette='magma')
ax[1].set_title('Top 5 Event Promosi dengan Pendapatan Terbesar', fontsize=12, fontweight='bold')
ax[1].set_xlabel('Total Pendapatan (Rupiah)')
ax[1].set_ylabel('Nama Event')

plt.tight_layout()
plt.show()

print("Perbandingan Rata-rata Nilai Transaksi (AOV):")
print(aov_promo)
```

```python
# Agregasi data berdasarkan metode pembayaran
metode_pembayaran_analisis = pemasukan_df.groupby('metode_pembayaran').agg(
    frekuensi_penggunaan=('id_transaksi', 'count'),
    total_nilai_transaksi=('total_harga', 'sum')
).reset_index()

# Visualisasi Perbandingan Frekuensi vs Total Nilai
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Grafik 1: Frekuensi Penggunaan (Popularitas)
sns.barplot(data=metode_pembayaran_analisis.sort_values(by='frekuensi_penggunaan', ascending=False),
            x='metode_pembayaran', y='frekuensi_penggunaan', ax=ax[0], palette='viridis')
ax[0].set_title('Metode Pembayaran Paling Sering Digunakan (Frekuensi)', fontsize=12, fontweight='bold')
ax[0].set_ylabel('Jumlah Transaksi')
ax[0].set_xlabel('Metode Pembayaran')

# Grafik 2: Total Nilai Transaksi (Nominal Uang)
sns.barplot(data=metode_pembayaran_analisis.sort_values(by='total_nilai_transaksi', ascending=False),
            x='metode_pembayaran', y='total_nilai_transaksi', ax=ax[1], palette='rocket')
ax[1].set_title('Metode Pembayaran dengan Nilai Transaksi Terbesar (Volume)', fontsize=12, fontweight='bold')
ax[1].set_ylabel('Total Nominal (Rupiah)')
ax[1].set_xlabel('Metode Pembayaran')

plt.tight_layout()
plt.show()

# Menampilkan dataframe hasil agregasi pembayaran
print("Tabel Analisis Metode Pembayaran:")
print(metode_pembayaran_analisis.sort_values(by='total_nilai_transaksi', ascending=False))
```

## Insight Exploratory Data Analysis (EDA)

Setelah melakukan eksplorasi mendalam terhadap data transaksi operasional UMKM periode 2024–2026, berikut adalah temuan insight yang berhasil diidentifikasi:

### 1. Analisis Tren Pendapatan
* Puncak Performa Bisnis: Pendapatan tertinggi UMKM selama 3 tahun terakhir tercapai pada bulan Mei 2025 sebesar Rp91,82 Juta dan Juli 2025 sebesar Rp91,10 Juta.

* Pola Musiman: Terdapat konsistensi di mana bulan Desember (baik di tahun 2025 maupun 2026) selalu masuk ke dalam jajaran 5 besar pendapatan tertinggi. Hal ini mengindikasikan adanya siklus lonjakan belanja konsumen di pertengahan tahun dan akhir tahun.

### 2. Analisis Kategori Produk
* Penggerak Utama Pendapatan: Kategori Fashion menjadi juara telak dengan menyumbang pendapatan sebesar Rp667,63 Juta. Menariknya, nilai fantastis ini diraih hanya dengan kuantitas penjualan yang relatif moderat (7.037 unit). Ini menandakan kategori Fashion memiliki margin atau harga satuan yang tinggi.

* Penggerak Perputaran Barang: Kategori Laundry (20.096 unit) dan Sembako (12.408 unit) menjadi kategori yang paling sering terjual, meskipun kontribusi pendapatannya berada di bawah Fashion. Ini adalah tipe produk kebutuhan harian yang menjaga arus kas operasional harian tetap berputar.

### 3. Analisis Event Promosi
* Anomali Nilai Keranjang: Terjadi anomali yang cukup unik di mana rata-rata nilai per transaksi tanpa promo justru sedikit lebih tinggi Rp74.955 dibandingkan transaksi menggunakan promo Rp72.295.

* Interpretasi Bisnis: Promo kemungkinan besar berhasil memicu konsumen untuk check-out lebih sering meningkatkan kuantitas/transaksi secara total, namun besaran diskon memotong nilai akhir keranjang belanja mereka. Strategi promo ke depan perlu dievaluasi agar menggunakan skema minimum purchase (misal: Diskon Rp10rb dengan minimal belanja Rp100rb) agar AOV ikut terdongkrak naik.

### 4. Analisis Metode Pembayaran
* Dominasi QRIS: Metode pembayaran QRIS mendominasi pasar UMKM ini secara mutlak, baik dari sisi popularitas maupun total perputaran uang.

* Cash & Transfer: Opsi Cash masih memegang peranan penting di posisi kedua sebesar Rp720 Juta, sementara Transfer Bank menjadi opsi yang paling jarang digunakan hanya 2.726 transaksi dengan total nominal Rp204 Juta. Integrasi sistem kasir digital yang lancar dengan QRIS wajib dipertahankan demi kenyamanan pelanggan belanja.

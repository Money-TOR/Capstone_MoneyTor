### Cleaning Data

```python
# Mengubah tipe data tanggal dari object menjadi datetime! #


# Kolom yang akan dikonversi antara lain
transactions_df['tanggal'] = pd.to_datetime(transactions_df['tanggal'])
sellers_df['tanggal_bergabung'] = pd.to_datetime(sellers_df['tanggal_bergabung'])
products_df['tanggal_input_produk'] = pd.to_datetime(products_df['tanggal_input_produk'])
logs_df['tanggal'] = pd.to_datetime(logs_df['tanggal'])

# Mengecek salah satu dataset yang sudah diubah kolom objectnya
transactions_df.info()
```

```python
# Menangani format kolom Diskon

# Menghapus teks pada kolom diskon
transactions_df['diskon'] = (
    transactions_df['diskon']
    .str.replace('Diskon ', '', regex=False)
    .str.replace('%', '', regex=False)
    .replace('Harga Normal', '0')
)

# Mengubah tipe data menjadi integer yang sebelumnya object
transactions_df['diskon'] = (
    transactions_df['diskon']
    .astype(int)
)

# Mengecek status kolom diskon pada dataset transactions
transactions_df[['diskon']].info()

# Melihat nilai unik kolom diskon setelah cleaning
print(transactions_df['diskon'].unique())
```

```python
# Validasi konsistensi perhitungan total_harga

# Menghitung ulang total transaksi
cek_total = transactions_df[
    (
        (
            transactions_df['qty']
            * transactions_df['harga_satuan']
            * (1 - transactions_df['diskon'] / 100)
        ).round(0)
        != transactions_df['total_harga']
    )
]

# Menampilkan hasil
cek_total
```

```python
# Mengecekk konsistensi Data Kategorikal

# memastikan tidak ada penulisan kategori yang berbeda
print("Metode Pembayaran:")
print(transactions_df['metode_pembayaran'].unique())

print("\nStatus Transaksi:")
print(transactions_df['status_transaksi'].unique())

print("\nJenis Transaksi:")
print(transactions_df['jenis_transaksi'].unique())

print("\nKategori:")
print(transactions_df['kategori'].unique())

print("\nEvent:")
print(transactions_df['event'].unique())

print("\nStatus Produk:")
print(products_df['status_produk'].unique())

print("\nJenis Perubahan:")
print(logs_df['jenis_perubahan'].unique())
```

```python
# Validasi log perubahan stok

# Memastikan perubahan stok masuk akal

cek_penjualan = logs_df[
    (logs_df['jenis_perubahan'] == 'Penjualan') &
    (
        logs_df['stok_sebelum']
        - logs_df['jumlah']
        != logs_df['stok_sesudah']
    )
]

cek_penjualan
```

## Exploratory Data Analysis (EDA)

### Hal yang diexplore adalah:

5. Distribusi transaksi berdasarkan waktu (tahunan dan bulanan)
6. Distribusi produk berdasarkan kategori dan status produk
7. Statistik harga dan stok produk
8. Karakteristik seller berdasarkan jenis usaha, lokasi, dan periode bergabung
9. Karakteristik supplier berdasarkan kategori dan lokasi supplier
10. Distribusi aktivitas inventori berdasarkan data log serta perubahan stok

Pada tahap ini dilakukan eksplorasi awal terhadap dataset UMKM untuk memahami karakteristik data `transaksi`, `produk`, `penjual`, `supplier`, dan aktivitas `inventori` selama periode 2024 - 2026. Eksplorasi dilakukan melalui pengamatan distribusi transaksi, metode pembayaran, pola waktu transaksi, karakteristik produk, kondisi stok, karakteristik seller dan supplier, serta aktivitas perubahan inventori.

Hasil eksplorasi ini digunakan untuk memperoleh pemahaman awal terhadap pola operasional UMKM dan hubungan antar data sebelum memasuki tahap Visualization & Explanatory Analysis untuk menjawab pertanyaan bisnis yang telah ditentukan.

```python
# Distribusi & Statistik Transaksi

print("===== Status Transaksi =====")
display(
    transactions_df['status_transaksi']
    .value_counts()
)

print("\n===== Kategori Transaksi =====")
display(
    transactions_df['kategori']
    .value_counts()
)

print("\n===== Metode Pembayaran =====")
display(
    transactions_df['metode_pembayaran']
    .value_counts()
)

print("\n===== Statistik Transaksi =====")
display(
    transactions_df[
        [
            'qty',
            'harga_satuan',
            'total_harga'
        ]
    ]
    .describe()
)
```

```python
# Distribusi Waktu Transaksi

transactions_df['tahun'] = (
    transactions_df['tanggal']
    .dt.year
)

transactions_df['bulan'] = (
    transactions_df['tanggal']
    .dt.month
)

print("===== Distribusi Tahunan =====")
display(
    transactions_df['tahun']
    .value_counts()
    .sort_index()
)

print("\n===== Distribusi Bulanan =====")
display(
    transactions_df['bulan']
    .value_counts()
    .sort_index()
)
```

```python
# Eksplorasi Kategori, Harga, dan Stok Produk

print("===== Distribusi Kategori Produk =====")
display(
    products_df['kategori_produk']
    .value_counts()
)

print("\n===== Distribusi Status Produk =====")
display(
    products_df['status_produk']
    .value_counts()
)

print("\n===== Statistik Harga dan Stok Produk =====")
display(
    products_df[
        [
            'harga_jual',
            'harga_modal',
            'stok_awal',
            'minimum_stok'
        ]
    ].describe()
)

print("\n===== 5 Produk dengan Harga Jual Tertinggi =====")
display(
    products_df[
        [
            'nama_produk',
            'kategori_produk',
            'harga_jual'
        ]
    ]
    .sort_values(
        by='harga_jual',
        ascending=False
    )
    .head()
)

print("\n===== 5 Produk dengan Stok Awal Tertinggi =====")
display(
    products_df[
        [
            'nama_produk',
            'kategori_produk',
            'stok_awal'
        ]
    ]
    .sort_values(
        by='stok_awal',
        ascending=False
    )
    .head()
)
```

```python
# Eksplorasi Karakteristik Seller
print("===== Distribusi Jenis Usaha =====")
display(
    sellers_df['jenis_usaha']
    .value_counts()
)

print("\n===== Distribusi Lokasi Usaha =====")
display(
    sellers_df['lokasi_usaha']
    .value_counts()
)

print("\n===== Periode Bergabung Seller =====")
display(
    sellers_df[
        [
            'nama_usaha',
            'tanggal_bergabung'
        ]
    ]
    .sort_values(
        by='tanggal_bergabung'
    )
)

print("\n===== Jumlah Seller =====")
print(
    sellers_df['id_seller']
    .nunique()
)
```

```python
# Eksplorasi Aktivitas Inventori

print("===== Distribusi Jenis Perubahan =====")
display(
    logs_df['jenis_perubahan']
    .value_counts()
)

print("\n===== Statistik Aktivitas Stok =====")
display(
    logs_df[
        [
            'jumlah',
            'stok_sebelum',
            'stok_sesudah'
        ]
    ].describe()
)

print("\n===== 5 Aktivitas dengan Perubahan Terbesar =====")
display(
    logs_df[
        [
            'id_produk',
            'jenis_perubahan',
            'jumlah'
        ]
    ]
    .sort_values(
        by='jumlah',
        ascending=False
    )
    .head()
)

print("\n===== 5 Stok Sesudah Tertinggi =====")
display(
    logs_df[
        [
            'id_produk',
            'stok_sesudah'
        ]
    ]
    .sort_values(
        by='stok_sesudah',
        ascending=False
    )
    .head()
)
```

```python
# Eksplorasi Supplier

print("===== Distribusi Kategori Supplier =====")
display(
    suppliers_df['kategori_supplier']
    .value_counts()
)

print("\n===== Distribusi Lokasi Supplier ===")
display(
    suppliers_df['lokasi']
    .value_counts()
)

print("\n===== Daftar Supplier =====")
display(
    suppliers_df[
        [
            'nama_supplier',
            'kategori_supplier',
            'lokasi'
        ]
    ]
)

print("\n===== Jumlah Supplier =====")
print(
    suppliers_df['id_supplier']
    .nunique()
)
```
